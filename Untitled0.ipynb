```python
import tokenize
from io import BytesIO
from collections import defaultdict

def categorize_tokens(source_code):
    # Convert source code to bytes
    source_bytes = BytesIO(source_code.encode('utf-8'))

    # Tokenize the source code
    tokens = tokenize.tokenize(source_bytes.readline)

    # Dictionaries to store categorized tokens
    categories = defaultdict(list)

    # Define token types
    token_types = {
        tokenize.OP: "operators",
        tokenize.NAME: "identifiers",
        tokenize.NUMBER: "numbers",
        tokenize.STRING: "strings",
        tokenize.NEWLINE: "newlines",
        tokenize.INDENT: "indentations",
        tokenize.DEDENT: "dedentations",
        tokenize.ENDMARKER: "endmarkers"
    }

    # Extract and categorize the tokens
    for token in tokens:
        if token.type in token_types:
            categories[token_types[token.type]].append(token.string)

    return categories

if __name__ == "__main__":
    # Example code to categorize tokens (Factorial Program)
    example_code = """
# Function to generate Fibonacci series
def fibonacci(n):
    fib_series = []
    a, b = 0, 1
    while len(fib_series) < n:
        fib_series.append(a)
        a, b = b, a + b
    return fib_series

# Number of terms
num_terms = 10  # You can change this to any number of terms you want

# Generate Fibonacci series
fib_series = fibonacci(num_terms)
print(f"Fibonacci series up to {num_terms} terms: {fib_series}")

"""

    categorized_tokens = categorize_tokens(example_code)

    for category, tokens in categorized_tokens.items():
        print(f"{category.capitalize()} are: {', '.join(tokens)}")
```

```python
from collections import defaultdict

class Graph:
    def __init__(self, graph):
        self.graph = graph
        self.rows = len(graph)

    def bfs(self, s, t, parent):
        visited = [False] * self.rows
        queue = []
        queue.append(s)
        visited[s] = True

        while queue:
            u = queue.pop(0)
            for ind, val in enumerate(self.graph[u]):
                if visited[ind] == False and val > 0:
                    queue.append(ind)
                    visited[ind] = True
                    parent[ind] = u
        return True if visited[t] else False

    def edmonds_karp(self, source, sink):
        parent = [-1] * self.rows
        max_flow = 0

        while self.bfs(source, sink, parent):
            path_flow = float("Inf")
            s = sink
            while s != source:
                path_flow = min(path_flow, self.graph[parent[s]][s])
                s = parent[s]

            max_flow += path_flow

            v = sink
            while v != source:
                u = parent[v]
                self.graph[u][v] -= path_flow
                self.graph[v][u] += path_flow
                v = parent[v]

        return max_flow

# Example network
graph = [
    [0, 8, 9, 0, 3, 0, 0, 0, 0],
    [0, 0, 0, 7, 2, 0, 3, 0, 0],
    [0, 0, 0, 0, 4, 3, 0, 6, 0],
    [0, 0, 0, 0, 0, 0, 4, 0, 0],
    [0, 0, 0, 0, 0, 0, 5, 3, 8],
    [0, 0, 0, 0, 0, 0, 0, 2, 0],
    [0, 0, 0, 0, 0, 0, 0, 0, 7],
    [0, 0, 0, 0, 0, 0, 0, 0, 7],
    [0, 0, 0, 0, 0, 0, 0, 0, 0]
]

g = Graph(graph)
source = 0
sink = 8

print("Max Flow:", g.edmonds_karp(source, sink))
```

```python
import numpy as np

# Define supply, demand, and cost matrices
supply = np.array([7, 9, 18])
demand = np.array([5, 8, 7, 14])
cost = np.array([[19, 30,50,10],
                 [70,30,40,60],
                 [40,8, 70,20, 999999]])  # Set a large integer for unallocated cells

# Initialize allocations matrix and total transportation cost
allocations = np.zeros((len(supply), len(demand)))
total_cost = 0

# Iterate until all supply and demand are satisfied
while np.sum(supply) > 0 and np.sum(demand) > 0:
    # Find the indices of the least cost cell
    min_cost_index = np.unravel_index(np.argmin(cost), cost.shape)
    i, j = min_cost_index

    # Calculate the quantity to allocate
    quantity = min(supply[i], demand[j])

    # Allocate the quantity
    allocations[i][j] = quantity

    # Update supply and demand
    supply[i] -= quantity
    demand[j] -= quantity

    # Calculate transportation cost for this allocation
    total_cost += quantity * cost[i][j]

    # Set cost to a large integer for cells with no supply/demand
    cost[i][j] = 999999

# Display the allocations matrix and total transportation cost
print("Allocations Matrix:")
print(allocations)
print("Total Transportation Cost:", total_cost)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
import numpy as np


def north_west_corner(costs, supply, demand):
    # Initialize empty allocation matrix
    allocation = np.zeros((len(supply), len(demand)))

    # Initialize indices for the north-west corner cell
    i, j = 0, 0

    while i < len(supply) and j < len(demand):
        # Find the minimum between supply[i] and demand[j]
        quantity = min(supply[i], demand[j])

        # Allocate the minimum quantity to the corresponding cell
        allocation[i, j] = quantity

        # Update supply and demand
        supply[i] -= quantity
        demand[j] -= quantity

        # Move to the next row or column if supply or demand is exhausted
        if supply[i] == 0:
            i += 1
        if demand[j] == 0:
            j += 1

    return allocation
costs = np.array([[2, 3, 11, 7], [1, 0, 6, 1], [5, 8, 15, 9]])
supply = [6, 1, 10]
demand = [7,5, 3, 2]

allocation = north_west_corner(costs, supply, demand)
print("Allocation Matrix:")
print(allocation)
```
